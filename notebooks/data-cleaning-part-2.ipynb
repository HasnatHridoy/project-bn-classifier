```python
allowed = set(labels)
```

```python
def find_illegal_str(lbl_str):
    if pd.isna(lbl_str):
        return True
    row_labels = [x.strip() for x in lbl_str.split(",")]
    for lbl in row_labels:
        if lbl not in allowed:
            return True
    return False

bad_rows = df[df["labels"].apply(find_illegal_str)]
```

```python
bad_rows
```

```python
invalid = set()

for row in bad_rows["labels"]:
    row_labels = [x.strip() for x in row.split(",")]
    for lbl in row_labels:
        if lbl not in allowed:
            invalid.add(lbl)

invalid
```

```python
len(invalid)
```

```python
def clean_label(labels_str):
    lbl_str = [x.strip() for x in labels_str.split(',')]
    clear_lbl = [l for l in lbl_str if l not in invalid]
    return ", ".join(clear_lbl)
```

```python
df['labels'] = df['labels'].apply(clean_label)
```

```python
df_after = df[df["labels"].apply(lambda lbls: any(l not in allowed for l in lbls))]
```

```python
bad_rows = df_after[df_after["labels"].apply(find_illegal_str)]
len(bad_rows)
```

```python
df = df_after.copy()
```

```python
def duplicate_free(lst):
    labels = [x.strip() for x in lst.split(',')]
    seen = set()
    result = []
    for l in labels:
        if l not in seen:
            result.append(l)
            seen.add(l)
    return ', '.join(result)
    
    
```

```python
df['labels'] = df['labels'].apply(duplicate_free)
```
