```python
import csv
import os
import time
import random
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException, WebDriverException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# ===== CONFIG =====
URLS_FILE = "/kaggle/input/notebook6f25afef66/all_links.csv"               
OUTPUT_FILE = "content_data_part(0-3000).csv"    
CHECKPOINT_FILE = "content_checkpoint_part1.txt"  
BATCH_SIZE = 100                           
WAIT_TIME = 10                             

START_INDEX = 0      
END_INDEX = 3000     

MAX_CONSECUTIVE_FAILS = 3

LONG_PAUSE_EVERY = 50       # every 50 pages, take a longer break
LONG_PAUSE_MIN = 120          
LONG_PAUSE_MAX = 180   
```

```python

# ===== UTILITY FUNCTIONS =====
def load_checkpoint():
    if os.path.exists(CHECKPOINT_FILE):
        with open(CHECKPOINT_FILE, "r") as f:
            return int(f.read().strip())
    return 0

def save_checkpoint(index):
    with open(CHECKPOINT_FILE, "w") as f:
        f.write(str(index))

def load_urls():
    with open(URLS_FILE, "r", encoding="utf-8") as f:
        return [line.strip() for line in f.readlines() if line.strip()]

def save_data(data, mode="a"):
    file_exists = os.path.exists(OUTPUT_FILE)
    with open(OUTPUT_FILE, mode, newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["title", "text", "labels", "url"])
        if not file_exists or mode == "w":
            writer.writeheader()
        writer.writerows(data)

# ===== SCRAPER FUNCTIONS =====
def scrape_content(driver, url):
    """Scrape title, article text, labels, and URL from a single page."""
    driver.get(url)
    wait = WebDriverWait(driver, WAIT_TIME)

    try:
        # title
        title_elem = wait.until(EC.presence_of_element_located(
            (By.CSS_SELECTOR, "h1.card-title.n_head")
        ))
        title = title_elem.text.strip()

        # article text
        article = wait.until(EC.presence_of_element_located((By.TAG_NAME, "article")))
        paragraphs = article.find_elements(By.TAG_NAME, "p")
        text = " ".join([p.text.strip() for p in paragraphs if p.text.strip()])

        # labels
        labels_area = driver.find_elements(By.CSS_SELECTOR, ".tagArea ul li a")
        labels = [a.text.strip() for a in labels_area if a.text.strip()]

        return {
            "title": title,
            "text": text,
            "labels": ", ".join(labels),
            "url": url
        }

    except TimeoutException:
        print(f"⚠️ Timeout while loading {url}")
        return None

def scrape_batch(driver, urls, start_index):
    """Scrape URLs in batches with checkpointing and error tolerance."""
    total = len(urls)
    fail_streak = 0
    scraped_data = []

    for i in range(start_index, total):
        url = urls[i]
        print(f"[{i+1}/{total}] Scraping: {url}")

        try:
            result = scrape_content(driver, url)
            if result:
                scraped_data.append(result)
                fail_streak = 0
            else:
                fail_streak += 1
        except WebDriverException as e:
            print(f"❌ Error scraping {url}: {e}")
            fail_streak += 1

        # Random small delay between requests
        time.sleep(random.uniform(1.5, 3))

        # Stop if too many consecutive failures
        if fail_streak >= MAX_CONSECUTIVE_FAILS:
            print("🚨 3 consecutive pages failed. Saving progress and stopping.")
            break

        # Save batch periodically
        if (i + 1) % BATCH_SIZE == 0 or i == total - 1:
            save_data(scraped_data)
            scraped_data.clear()
            save_checkpoint(i + 1)
            print(f"✅ Batch saved at index {i + 1}")
            
            if (i + 1) % LONG_PAUSE_EVERY == 0:
                sleep_time = random.uniform(LONG_PAUSE_MIN, LONG_PAUSE_MAX)
                print(f"⏸ Taking a longer pause for {sleep_time:.1f} seconds...")
                time.sleep(sleep_time)

    # Save any remaining data
    if scraped_data:
        save_data(scraped_data)
        save_checkpoint(i + 1)
        print("✅ Final batch saved.")

# ===== MAIN =====
def main(driver):
    # Load URLs and slice for this resource
    urls = load_urls()[START_INDEX:END_INDEX]
    start_index = load_checkpoint()
    print(f"Starting from index {start_index} (resource range {START_INDEX}-{END_INDEX})")

    scrape_batch(driver, urls, start_index)
    print("🎯 Scraping complete!")

```

```python
main(driver)
```
