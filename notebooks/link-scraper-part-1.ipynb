```python
import os
import csv
import time
import random

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python


# -------------------------
# CONFIG
# -------------------------
START_PAGE = 1
END_PAGE = 1042
BASE_URL = "https://www.bd-pratidin.com/national?page={}"
CHECKPOINT_FILE = "links_checkpoint.csv"

MAX_RETRIES = 3
WAIT_BETWEEN_RETRIES = 5
BATCH_SAVE_EVERY = 20
MAX_CONSECUTIVE_FAILS = 3  


# -------------------------
# LOAD CHECKPOINT
# -------------------------
collected = set()
last_page = START_PAGE - 1  # default to start

if os.path.exists(CHECKPOINT_FILE):
    with open(CHECKPOINT_FILE, newline="", encoding="utf-8") as f:
        reader = csv.reader(f)
        for row in reader:
            if row:
                if row[0] == "__LAST_PAGE__":
                    last_page = int(row[1])
                else:
                    collected.add(row[0])

print(f"Resuming from page {last_page + 1}, {len(collected)} links collected")

# -------------------------
# UTILITY FUNCTIONS
# -------------------------
def save_checkpoint(urls, current_page):
    mode = "a" if os.path.exists(CHECKPOINT_FILE) else "w"
    with open(CHECKPOINT_FILE, mode, newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        for u in urls:
            writer.writerow([u])
        writer.writerow(["__LAST_PAGE__", current_page])

def safe_get(driver, url):
    for attempt in range(1, MAX_RETRIES + 1):
        try:
            driver.get(url)
            return True
        except Exception as e:
            print(f"[attempt {attempt}] Failed to load {url}: {e}")
            if attempt < MAX_RETRIES:
                time.sleep(WAIT_BETWEEN_RETRIES * attempt)
            else:
                return False

# -------------------------
# SCRAPING LOOP
# -------------------------
new_batch = []
consecutive_fail_count = 0

try:
    for page_num in range(last_page + 1, END_PAGE + 1):
        url = BASE_URL.format(page_num)
        if not safe_get(driver, url):
            print(f"[page {page_num}] failed to load")
            consecutive_fail_count += 1
            if consecutive_fail_count > MAX_CONSECUTIVE_FAILS:
                print("Too many consecutive failures. Shutting down.")
                break
            continue

        # wait for container
        try:
            WebDriverWait(driver, 12).until(
                EC.presence_of_element_located((By.CLASS_NAME, "col-12"))
            )
        except:
            print(f"[page {page_num}] container not found")
            consecutive_fail_count += 1
            if consecutive_fail_count > MAX_CONSECUTIVE_FAILS:
                print("Too many consecutive failures. Shutting down.")
                break
            continue

        # reset fail counter on success
        consecutive_fail_count = 0

        # collect links
        try:
            container = driver.find_element(By.CLASS_NAME, "col-12")
            stretched = container.find_elements(By.CSS_SELECTOR, "a.stretched-link")
        except:
            stretched = []

        added = 0
        for a in stretched:
            href = a.get_attribute("href")
            if not href:
                continue
            href = href.strip()
            if href not in collected:
                collected.add(href)
                new_batch.append(href)
                added += 1

        print(f"[page {page_num}] found {len(stretched)} links, {added} new")

        # checkpoint/save periodically
        if len(new_batch) >= BATCH_SAVE_EVERY:
            save_checkpoint(new_batch, page_num)
            new_batch = []

        # polite random wait
        time.sleep(random.uniform(2.5, 5))

        # optional: longer pause every 200 pages
        if page_num % 100 == 0:
            print(f"[info] longer break at page {page_num}")
            time.sleep(random.uniform(60, 120))

finally:
    # save remaining
    if new_batch:
        save_checkpoint(new_batch, page_num)
    driver.quit()
    print(f"Scraping finished. Total links collected: {len(collected)}")

# -------------------------
# CLEAN FINAL CSV
# -------------------------
final_file = "all_links.csv"
with open(CHECKPOINT_FILE, newline="", encoding="utf-8") as f_in, \
     open(final_file, "w", newline="", encoding="utf-8") as f_out:
    writer = csv.writer(f_out)
    for row in csv.reader(f_in):
        if row and row[0] != "__LAST_PAGE__":
            writer.writerow([row[0]])

print(f"Final clean links saved to {final_file}")
```
